**1. What is the difference between a function and a method in Python?**
- A function in Python is an independent block of code that performs a specific task and can be called by its name from anywhere in the program, while a method is a function that is associated with an object or class and is called on that object using dot notation.


```python
# Function Example:
def greet(name):
    return f"Hello, {name}!"

print(greet("Vijay"))  # Calling a function
```

```python
# Method Example:

class Person:
    def greet(self):  # This is a method
        return f"Hello, I am a person!"

p = Person()
print(p.greet())  # Calling a method
```

**4. What is the purpose of the `return` statement in a Python function?**
- The return statement in a Python function serves to end the function’s execution and send a value back to the part of the program that called the function.

**5.  What are iterators in Python and how do they differ from iterables?**
 - In Python, an iterator is an object used to traverse through all the elements of a collection, one at a time, using the __next__() method. An iterable is any object capable of returning its elements, ideally one at a time, and can be passed to the built-in iter() function to obtain an iterator.

 - Iterators:
   - An iterator is an object that keeps track of where you are while iterating.
   - It is produced by calling iter() on an iterable.
   - It has two key methods:
     - __iter__() (returns itself)
     - __next__() (returns the next item, raises StopIteration when done)
 - Iterables
   - An iterable is any Python object capable of returning its members one at a time.
   - You can loop over it with a for loop or pass it to functions like iter().
   - Examples: list, tuple, str, dict, set, range()…

```python
# Example of Iterators
my_list = [1, 2, 3]         # iterable
it = iter(my_list)          # create an iterator from the iterable

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
# next(it) now raises StopIteration

```

```python
# Example of Iterable

my_list = [1, 2, 3]  # This is an iterable
for item in my_list:  # We can loop over it
    print(item)
```

**6.  Explain the concept of generators in Python and how they are defined.**
- Generators in Python are special types of functions that return an iterator and produce a sequence of values one at a time, pausing their state after yielding each value and resuming from there on the next call. Unlike regular functions that use return to send back a single value and terminate, generators use the yield statement to produce a series of values over time without losing their execution context.
- How Generators are Defined :
  - Generator functions use the def keyword like normal functions but include one or more yield statements instead of return.
  - When called, these functions return a generator object that can be iterated over to produce values on-demand.

```python
# Example

def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(5):
    print(num)
```

**7. What are the advantages of using generators over regular functions?**
- The advantages of using generators over regular functions in Python primarily relate to memory efficiency, lazy evaluation, and handling large or infinite data streams gracefully.
- Advantages of Generators
  - Memory Efficiency: Generators produce items one at a time and only when needed, so they do not store the entire sequence in memory unlike regular functions that might return a whole list or collection at once.
  - Lazy Evaluation: They compute values on-the-fly as required, saving processing time and resources if not all values are needed immediately or ever.
  - Ability to Handle Infinite Sequences: Generators can represent endless sequences without running into memory issues since they yield results incrementally.
  - Maintains State Between Yields: Generators pause execution upon yielding a value and resume efficiently later without losing context, which regular functions cannot do.
  - Improved Performance with Large Datasets: When working with large amounts of data, generators can be much faster and more responsive since they don’t create large intermediate data structures.
  - Cleaner Code for Iteration: Generators can simplify code for complex iterative processes, avoiding the need for manual bookkeeping of iteration state.

```python
# Example

def square_gen(n):
    for i in range(n):
        print(f"Generating {i}")
        yield i * i

for val in square_gen(3):
    print(val)

```

**9. Explain the purpose and usage of the `map()` function in Python.**
- The map() function in Python applies a specified function to each item of one or more iterables (like lists, tuples) and returns a map object (an iterator) of the results.
- Purpose
  - To transform or process each element in an iterable by applying a function to it.
  - Useful for applying the same operation to many items without writing explicit loops.

```python
def square(x):
    return x * x;

numbers = [1, 2, 3, 4]
squared = map(square, numbers)
print(list(squared))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
- Map
  - purpose: Applies a given function to each element of one or more iterables and returns an iterator of the results
  - How it Works: Transforms each element individually by applying a function
  - Example Use Case : Doubling each number in a list
- Reduce
  - purpose: Applies a function cumulatively to items of an iterable to reduce it to a single value
  - How it Works: Combines elements pairwise to produce a summary result
  - Example Use Case : Calculating the product or sum of list elements

- Filter
  - purpose:Filters elements from an iterable by applying a function that returns True or False; returns only those elements for which the function returns True
  - How it Works: Selects elements based on a condition
  - Example Use Case : Selecting even numbers from a list

```python
# Example = Map
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
```

```python
# Example = Reduce

from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
```

```python
# Example = Filter

numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.**

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(nums)
print(result)
```

```python
def square_list(numbers):
    """
    This function takes a list of integers and returns
    a new list containing the squares of each number.
    """
    squared_numbers = [num**2 for num in numbers]  # list comprehension
    return squared_numbers

# Example usage
my_list = [1, 2, 3, 4, 5]
squared_list = square_list(my_list)
print("Squared list:", squared_list)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    """
    This function checks if n is a prime number.
    Returns True if prime, False otherwise.
    """
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # only check up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Example: Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")

```

**5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n  # number of terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self  # an iterator returns itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # stop when we reach the specified number of terms
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            fib = self.a + self.b
            self.a, self.b = self.b, fib
            return fib

# Example usage: Generate first 10 Fibonacci numbers
fib_sequence = Fibonacci(10)
for num in fib_sequence:
    print(num, end=" ")
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent**

```python
def powers_of_two(max_exponent):
    """
    Generator function that yields powers of 2 from 0 to max_exponent.
    """
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage: powers of 2 up to 10
for value in powers_of_two(10):
    print(value, end=" ")
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line
    and yields each line as a string.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip("\n")  # strip newline at the end

# Example usage:
# Save a sample text file first
with open("sample.txt", "w") as f:
    f.write("First line\nSecond line\nThird line")

# Use the generator
for line in read_file_line_by_line("sample.txt"):
    print(line)


```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to convert each temperature
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_string = "Hello World"

# Use filter() to remove vowels
result = ''.join(filter(is_not_vowel, input_string))

print("Original string:", input_string)
print("Without vowels:", result)
```
